## Nearest Neighbor Resampling

### Evapotranspiration - Resampling Only

```python
import xarray as xr
import numpy as np
ds = xr.open_dataset(r"D:\Projects\GNR644 - Water Balance\Evapotranspiration\GLEAM_Evapotranspiration_cropped.nc")

new_lat = np.arange(ds.lat.min(), ds.lat.max(), 0.25)
new_lon = np.arange(ds.lon.min(), ds.lon.max(), 0.25)

ds_025 = ds.interp(lat=new_lat, lon=new_lon, method="nearest")

ds_025.to_netcdf(r"D:\Projects\GNR644 - Water Balance\Evapotranspiration\GLEAM_Evapotranspiration_resampled_NN.nc")
```

### Groundwater (Resampling and Cropping)

```python
import numpy as np
```

```python
ds = xr.open_dataset(r"D:\Projects\GNR644 - Water Balance\Groundwater\GRCTellus.JPL.200204_202510.GLO.RL06.3M.MSCNv04CRI.nc")

ds = ds.rio.write_crs("EPSG:4326")
ds = ds.rio.set_spatial_dims(x_dim="lon", y_dim="lat")

lwe = ds["lwe_thickness"]

new_lat = np.arange(float(lwe.lat.min()), float(lwe.lat.max()), 0.25)
new_lon = np.arange(float(lwe.lon.min()), float(lwe.lon.max()), 0.25)

# After nearest neighbor interpolation
lwe_025 = lwe.interp(
    lat=new_lat,
    lon=new_lon,
    method="nearest"
)

# reattach CRS & spatial dims here
lwe_025 = lwe_025.rio.write_crs("EPSG:4326")
lwe_025 = lwe_025.rio.set_spatial_dims(x_dim="lon", y_dim="lat")


gdf = gpd.read_file(r"D:\Projects\GNR644 - Water Balance\PeninsularIndia\PeninsularIndia\peninsularIndia_lvl3.shp")
gdf = gdf.to_crs("EPSG:4326")

lwe_clip = lwe_025.rio.clip(gdf.geometry, gdf.crs, drop=True)
lwe_clip.isel(time=0).plot()

lwe_clip.to_netcdf(r"D:\Projects\GNR644 - Water Balance\Groundwater\Groundwater_resample_crop.nc")
```

### Precipitation (Resampling and Cropping)

```python
ds = xr.open_dataset(r"D:\Projects\GNR644 - Water Balance\Precipitation\GPCP_2000_2024_Monthly.nc")

precip = ds["satellite_precip"]
new_lat = np.arange(float(precip.lat.min()), float(precip.lat.max()), 0.25)
new_lon = np.arange(float(precip.lon.min()), float(precip.lon.max()), 0.25)

precip_025 = precip.interp(
    lat=new_lat,
    lon=new_lon,
    method="nearest"   
)
precip_025 = precip_025.rio.write_crs("EPSG:4326")
precip_025 = precip_025.rio.set_spatial_dims(x_dim="lon", y_dim="lat")
gdf = gpd.read_file(r"D:\Projects\GNR644 - Water Balance\PeninsularIndia\PeninsularIndia\peninsularIndia_lvl3.shp")
gdf = gdf.to_crs("EPSG:4326")
precip_clip = precip_025.rio.clip(gdf.geometry, gdf.crs, drop=True)
precip_clip.isel(time=0).plot()
precip_clip.to_netcdf(r"D:\Projects\GNR644 - Water Balance\Precipitation\Rainfall_Resample_Cropped.nc")
```

### Runoff (Cropping)

```python
import xarray as xr
import geopandas as gpd
import rioxarray

# Open dataset
ds = xr.open_dataset("/content/drive/MyDrive/GNR644 - Water Balance/Data/Runoff/Runoff.nc")

# Assign CRS and spatial dimensions
ds = ds.rio.write_crs("EPSG:4326")
ds = ds.rio.set_spatial_dims(x_dim="longitude", y_dim="latitude")

# Read shapefile
gdf = gpd.read_file("/content/drive/MyDrive/GNR644 - Water Balance/Data/PeninsularIndia/peninsularIndia_lvl3.shp")
gdf = gdf.to_crs("EPSG:4326")

# Clip directly (NO resetting of spatial dims here)
clipped_RO = ds["ro"].rio.clip(gdf.geometry, gdf.crs, drop=True)

# Plot first timestep
clipped_RO.isel(valid_time=0).plot()

# Save
clipped_RO.to_netcdf("/content/drive/MyDrive/GNR644 - Water Balance/Data/Runoff/Runoff_cropped.nc")
```

### Soil Moisture (Cropping)

```python
import xarray as xr
import geopandas as gpd
import rioxarray

ds = xr.open_dataset("/content/drive/MyDrive/GNR644 - Water Balance/Data/Soil Moisture/SoilMoisture_2000_2024_Monthly.nc")
print(ds.coords)

ds = ds.rio.write_crs("EPSG:4326")  # lat/lon WGS84
ds = ds.rio.set_spatial_dims(x_dim="lon", y_dim="lat")

gdf = gpd.read_file("/content/drive/MyDrive/GNR644 - Water Balance/Data/PeninsularIndia/peninsularIndia_lvl3.shp")
gdf = gdf.to_crs("EPSG:4326")  # must match NetCDF CRS

# Select the main data variable ('satellite_precip') and explicitly set its spatial dimensions before clipping
clipped_ET = ds["sm"].rio.write_crs("EPSG:4326").rio.set_spatial_dims(x_dim="lon", y_dim="lat").rio.clip(gdf.geometry, gdf.crs, drop=True)

# Plotting the clipped variable
clipped_ET.isel(time=0).plot()

# Save the clipped variable to a new NetCDF file
clipped_ET.to_netcdf("/content/drive/MyDrive/GNR644 - Water Balance/Data/Evapotranspiration/GLEAM_Evapotranspiration_crop.nc")
```
